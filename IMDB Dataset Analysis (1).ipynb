# 20. Classify movies Based on Ratings[Excellent, Good and Average]

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
        
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

# 22. Find unique value from the genre

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```

# 23. How many films of each genre were made?

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
from collections import Counter
```

```python
Counter(one_d)
```

```python
genre_counts = {genre: one_d.count(genre) for genre in uni_list}
genre_series = pd.Series(genre_counts).sort_values(ascending=False)
plt.figure(figsize=(14, 12))
plt.pie(genre_series, labels=genre_series.index, autopct='%1.1f%%', startangle=170, colors=plt.cm.Paired.colors)
plt.title('Distribution of Movie Genres')
plt.show()
```
